# Análisis de Datos de Compra de Productos
Este notebook presenta un análisis exploratorio de un pequeño conjunto de datos de clientes y sus compras, utilizando pandas y matplotlib para visualizar patrones.

```python
# Importamos las bibliotecas necesarias
import pandas as pd  # Para manipular datos tabulares
import matplotlib.pyplot as plt  # Para crear gráficos
```

```python
# Creamos un pequeño dataset con datos de ejemplo
datos = {
    'edad': [25, 45, 35, 33, 50, 23, 40, 60],
    'ingresos': [2000, 5000, 3500, 3000, 7000, 1800, 4200, 8000],
    'compró': [0, 1, 1, 0, 1, 0, 1, 1]
}
df = pd.DataFrame(datos)  # Convertimos el diccionario en un DataFrame
df
```

```python
# Contamos cuántos clientes compraron y cuántos no
conteo_compras = df['compró'].value_counts()
print("Conteo de compras (0 = No, 1 = Sí):\n", conteo_compras)
```

```python
# Visualizamos el conteo de compras en un gráfico de barras
conteo_compras.plot(kind='bar', color=['red', 'green'])
plt.title('Distribución de Compras')
plt.xlabel('Compró (0 = No, 1 = Sí)')
plt.ylabel('Número de Clientes')
plt.xticks([0, 1], ['No', 'Sí'], rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Visualizamos la relación entre edad e ingresos
plt.figure(figsize=(8, 5))
plt.scatter(df['edad'], df['ingresos'], c=df['compró'], cmap='bwr', edgecolor='k')
plt.title('Relación entre Edad e Ingresos')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.colorbar(label='Compró (0 = No, 1 = Sí)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
